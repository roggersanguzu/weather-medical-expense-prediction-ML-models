```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',150)


sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10,6)
matplotlib.rcParams['patch.edgecolor']='#313338'
```

```python
link='https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package/data'
od.download(link)
path='/content/weather-dataset-rattle-package'
os.listdir(path)
raw_df=pd.read_csv(path+'/weatherAUS.csv')
raw_df
```

```python
a=raw_df.copy()
b=raw_df.copy()
c=raw_df.copy()
d=raw_df.copy()
```

```python
a.dropna(subset=['RainToday','RainTomorrow'],inplace=True)
a.info()
```

```python
use_sample=True
sample_fraction=0.1

if use_sample:
  a=a.sample(frac=sample_fraction).copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df,test_df=train_test_split(a,test_size=0.2,random_state=42)
train_df,val_df=train_test_split(train_val_df,test_size=0.25,random_state=42)
print('train_val_df',train_val_df.shape)
print('test_df',test_df.shape)
print('train_df',train_df.shape)
print('val_df',val_df.shape)
```

```python
year=pd.to_datetime(a.Date).dt.year
train_df=a[year<2015]
test_df=a[year>=2015]
val_df=a[year==2015]
```

```python
inputcolumns=list(train_df.columns)[1:-1]
targetcolumns='RainTomorrow'
inputcolumns
```

```python
train_inputs=train_df[inputcolumns].copy()
train_targets=train_df[targetcolumns].copy()

val_inputs=val_df[inputcolumns].copy()
val_targets=val_df[targetcolumns].copy()

test_inputs=test_df[inputcolumns].copy()
test_targets=test_df[targetcolumns].copy()

train_inputs
```

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols=train_inputs.select_dtypes(include='object').columns.tolist()
```

```python
train_inputs[numeric_cols].describe()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
imputer.fit(a[numeric_cols])
```

```python
train_inputs[numeric_cols]=imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=imputer.transform(test_inputs[numeric_cols])
test_inputs
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(a[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(handle_unknown='ignore')
encoder.fit(a[categorical_cols])
```

```python
encoded_cols=list(encoder.get_feature_names_out(categorical_cols))
print(encoded_cols)
```

```python
train_encoded = encoder.transform(train_inputs[categorical_cols].fillna('unknown'))
val_encoded = encoder.transform(val_inputs[categorical_cols].fillna('unknown'))
test_encoded = encoder.transform(test_inputs[categorical_cols].fillna('unknown'))

train_inputs[encoded_cols] = train_encoded.toarray()
val_inputs[encoded_cols] = val_encoded.toarray()
test_inputs[encoded_cols] = test_encoded.toarray()
```

```python
x_train=train_inputs[numeric_cols + encoded_cols]
x_val=val_inputs[numeric_cols + encoded_cols]
x_test=test_inputs[numeric_cols + encoded_cols]
```

DECISION TREES IN AI&ML

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(random_state=42)
model.fit(x_train,train_targets)
```

```python
train_pred=model.predict(x_train)
print(train_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(train_pred,train_targets)
```

```python
confusion_matrix(train_pred,train_targets)
```

predictions can be made and predictions made at the same time

```python
model.score(x_val,val_targets)
```
