```python
# Roggers Anguzu.
# Using multiple features
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import seaborn as sns
import plotly.express as px
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/roggersanguzu/Machine-Learning-with-R-datasets/master/insurance.csv"
medicdata = pd.read_csv(url)
```

```python
sns.set_style("darkgrid")
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10,5)
```

```python
batch1=medicdata.copy()
batch2=medicdata.copy()
batch3=medicdata.copy()
batch4=medicdata.copy()
```

```python
batch1['smoker'] = batch1['smoker'].str.strip().str.lower()
batch1['smoker'] = batch1['smoker'].map({'yes': 1, 'no': 0})
batch1
```

```python
smokers=batch1[batch1['smoker']==1]
nonsmoker=batch1[batch1['smoker']==0]
```

```python
def estimate(age,m,c):
  return m*age + c
```

```python
def try_estimates(m,c):
  ages=nonsmoker.age
  target=nonsmoker.charges
  estimated_charges = estimate(ages,m,c)

  plt.plot(ages,estimated_charges,'r')
  plt.scatter(x=ages,y=target)
  plt.xlabel=('age')
  plt.ylabel=('charges')
  plt.legend(['Estimate','Actual'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error as rmse
input,target=nonsmoker[['age','bmi']],nonsmoker[['charges']]
model=LinearRegression().fit(input,target)
```

```python
predictions=model.predict(input)
predictions
```

```python
rmse(target,predictions)
```
